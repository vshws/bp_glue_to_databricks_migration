```python
# ==============================
# STEP 1: Read from S3 (raw)
# ==============================

df = spark.read.option("header", "true") \
    .csv("s3://bp-glue-demo/raw/counterparties/")

display(df)
```

```python
# ==============================
# STEP 2: Write to Bronze table
# ==============================

df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("udp.bronze.counterparties")
```
